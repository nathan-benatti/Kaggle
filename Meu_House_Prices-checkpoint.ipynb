## Missing Data Treatment

We will check for missing data on features, the infamous NaN values, and see what we will do with them

```python
# Checking for missing values and sorting values in descending order
sb.set_style("whitegrid")
missing_train = train.isnull().sum()
missing_train = missing_train[missing_train>0]
missing_train.sort_values(ascending=False,inplace=True)
missing_train.plot.bar(figsize=(12,5))
```

```python
# comparing missing data with, the missing data/number_of_samples coeficient
percent = (train.isnull().sum()/train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([missing_train,percent],axis=1,keys=['Missing','Percent'])
missing_data.head(20)
```

```python
# comparing missing data with, the missing data/number_of_samples coeficient, for test dataset
missing_test = test.isnull().sum()
missing_test = missing_test[missing_test>0]
percent = (test.isnull().sum()/test.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([missing_test,percent],axis=1,keys=['Missing','Percent'])
missing_data.head(40)
```

We are sure missing a lot of pools here, maybe because pools are just for some rich bastards !! But, being serious now, i will adopt the convenience that if we are missing more than 15% of the feature data, we will drop it, cause we can't have much use of it.

```python
# Usarei a convenÃ§Ã£o de dispensar variaveis que em mais de 15% das amostras tem seus dados perdidos
train.drop(['PoolQC','MiscFeature','Alley','Fence','FireplaceQu','LotFrontage'],axis=1,inplace=True)
test.drop(['PoolQC','MiscFeature','Alley','Fence','FireplaceQu','LotFrontage'],axis=1,inplace=True)
```

Now i will try to analyse features with missing data that have correlation with better features, so i can drop the worst one, if i find them redundant.Checking the heatmap of correlations i came to some conclusions:
As said earlier about GarageYrBlt and YrBlt correlation, i will drop GarageYrBlt considering YrBlt  sufficient.As for others Garage related features,i will drop every one of them, except GarageCars and Garage Area, for those are suficiently correlated to support the Garage importance. But, considering the two redundant, i will just stick with GarageCars feature.

```python
train.drop(['GarageYrBlt','GarageType','GarageFinish','GarageQual','GarageCond','GarageArea','BsmtFinType2','BsmtExposure',
'BsmtFinType1','BsmtCond','BsmtQual','BsmtFullBath','BsmtHalfBath','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF'],axis=1,inplace=True)
test.drop(['GarageYrBlt','GarageType','GarageFinish','GarageQual','GarageCond','GarageArea','BsmtFinType2','BsmtExposure',
'BsmtFinType1','BsmtCond','BsmtQual','BsmtFullBath','BsmtHalfBath','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF'],axis=1,inplace=True)
```

For all other missing data, i will apply the mean value for the quantitative features, and the mode value for the qualitative features:

```python
print(train['MasVnrArea'].dtype)
print(train['MasVnrType'].dtype)
print(train['Electrical'].dtype)
print('\n')
print(test['MSZoning'].dtype)
print(test['GarageCars'].dtype)
print(test['KitchenQual'].dtype)
print(test['TotalBsmtSF'].dtype)
print(test['Utilities'].dtype)
print(test['Functional'].dtype)
print(test['Exterior1st'].dtype)
print(test['Exterior2nd'].dtype)
print(test['SaleType'].dtype)
```

```python
for col in ('MasVnrArea','GarageCars','TotalBsmtSF'):
    train[col]=train[col].fillna(train[col].mean())
    test[col]=test[col].fillna(test[col].mean())
```

## Variable Transformation

To apply the prediction models, all my data have to be numerical, so we will encode with get_dummies function the categorical features

```python
train_feat = train.drop('SalePrice',axis=1) # Just want the features, droping the saleprice column
test_feat = test

trainlen = len(train_feat) # saving train length
```

```python
dataset = pd.concat(objs=[train_feat,test_feat],axis=0).reset_index(drop=True) # meshing the datasets, important to keep the same columns number
dataset = pd.get_dummies(dataset) # coding categorical features to numerical

train_feat = dataset[:trainlen] # getting trainset back
test_feat = dataset[trainlen:]  # getting testset back
```

## Modelling
 For modelling porpuse , i will make a train_test_split, because this test dataframe we don't have the SalePrice column for competition issues, so i will create other test dataframe from the train dataset, and i will called them SUBtrain and SUBtest, just so it don't become too confusing

```python
xtrain = train_feat 
SalePricelog = np.log(train['SalePrice']) # log transform to normalize the distribution
ytrain = SalePricelog
```

```python
from sklearn.model_selection import train_test_split

subX_train,subX_test,subY_train,subY_test = train_test_split(xtrain,ytrain,test_size=0.2)
```

```python
import xgboost as xgb
```

### XGBoost Hyper Tunning Parameters

```python
import xgboost as xgb
model = xgb.XGBRegressor() # using default parameters
model.fit(subX_train,subY_train)

print('(Sub)train accuracy: ', model.score(subX_train,subY_train))
print('(Sub)test accuracy: ', model.score(subX_test,subY_test))
```

Train accuracy is almost perfect, but doesnt fit quite well in test set, with only 88% accuracy of the test dataframe, showing us the model is probably overfit, so what i will try to tune the parameters to fix this.

I will tune by this fine gentleman method here : https://www.youtube.com/watch?v=AvWfL1Us3Kg&t=1476s , in case you are interessed.

```python
model = xgb.XGBRegressor(max_depth=4,
                         subsample=0.8,
                         n_estimators=400,
                         learning_rate=0.05,
                         min_child_weight=2,
                         colsample_bylevel=0.4,
                         
                        )
model.fit(subX_train,subY_train)

print('(Sub)train accuracy: ', model.score(subX_train,subY_train))
print('(Sub)test accuracy: ', model.score(subX_test,subY_test))
```

Tuning the parameters i could fix a lit bit the overfit, and the test accuracy raise almost 2%.
